```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
r0 = 1
omega = 0.006* np.pi
v0 = 0.005
phi0 = 0

def r(t):
    return r0*np.cosh(omega*t) + v0/omega*np.sinh(omega*t)

def phi(t):
    return omega*t + phi0

def rphi(phi):
    return r0*np.cosh(phi-phi0) + v0/omega * np.sinh(phi-phi0)
```

```python
t = np.linspace(0, 60, 1000)
phivalues = phi(t)
```

```python
plt.plot(t, r(t))
plt.xlabel('t (s)')
plt.ylabel('r (m)')
plt.title(r'Gráfica para $r(t)$')
```

```python
plt.plot(t, phi(t))
plt.xlabel('t (s)')
plt.ylabel(r'$\phi$ (rad)')
plt.title(r'Gráfica para $phi(t)$')
```

```python
plt.axes(projection = 'polar') 
plt.polar(phivalues, rphi(phivalues))
plt.title(r'Gráfica para $r(\phi)$')
```
